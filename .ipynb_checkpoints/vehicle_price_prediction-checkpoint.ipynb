```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('Car details v3.csv')
df.head()
```

```python
df.shape
```

```python
df.duplicated().any()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.info()
```

I decided to remove the torque column since it is too technical and specific, and it is also rated at different rpm, hence it is not providing much information.

```python
df.drop(['torque'],axis=1, inplace = True)
df.head()
```

Check for missing values

```python
df.isnull().any()
```

```python
df.isnull().sum() / df.shape[0] * 100
```

Since we have more than 6000 samples which is decent, I decided to remove the rows with missing values since they only constitute about 3% of the data.

```python
df.dropna(axis=0, inplace=True)
df.isnull().any()
```

```python
df.shape
```

```python
df['owner'] = df['owner'].replace({'First Owner': 1, 'Second Owner': 2, 'Third Owner': 3})
df.head()
```

```python
df['mileage'] = df['mileage'].str.strip('kmpl')
df['engine'] = df['engine'].str.strip('CC')
df['max_power'] = df['max_power'].str.strip('bhp')
df.head()
```

Let's do some exploratory analysis

```python
trace2 = go.Box(
    y=df['selling_price'],
    name = 'selling_price',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
data = [trace2]
layout = go.Layout(go.Layout(title="Selling price in Rupee"))
fig = go.Figure(data, layout=layout)
fig.show()
```

```python
trace1 = go.Box(
    y=df['km_driven'],
    name = 'km_driven',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
data = [trace1]
layout = go.Layout(go.Layout(title="Total km driven"))
fig = go.Figure(data, layout=layout)
fig.show()
```

```python
count_fuel = df['fuel'].value_counts().reset_index()
count_fuel = count_fuel.rename(columns = {'index':'fuel','fuel':'count'})
```

```python

trace1 = go.Bar(y=count_fuel['count'], x=count_fuel['fuel'], name="")

data = [trace1]
layout = go.Layout(title="Fuel type",xaxis_title='fuel',yaxis_title='count', height=700, legend=dict(x=0.1, y=1.1))
fig = go.Figure(data, layout=layout)
fig.show()
```
